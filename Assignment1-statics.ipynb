```python
import pandas as pd
import numpy as np
df = pd.read_csv('house_price.csv')
print(df.head())
print(df.describe())
# Check for missing values
print(df.isnull().sum())
print(df.dtypes)
```

```python
mean = df['price_per_sqft'].mean()
print(mean)
```

```python
std_dev = df['price_per_sqft'].std()
print(std_dev)
```

```python
lower_limit = mean - 3 * std_dev
```

```python
print(lower_limit)
```

```python
upper_limit = mean + 3 * std_dev
```

```python
print(upper_limit)
```

```python

df_cleaned_mean_std = df[(df['price_per_sqft'] >= lower_limit) & (df['price_per_sqft'] <= upper_limit)]
print(df_cleaned_mean_std)
```

```python
lower_percentile = df['price_per_sqft'].quantile(0.01)
upper_percentile = df['price_per_sqft'].quantile(0.99)
print(lower_percentile)
print(upper_percentile)
df_cleaned_percentile = df[(df['price_per_sqft'] >= lower_percentile) & (df['price_per_sqft'] <= upper_percentile)]
print(df_cleaned_percentile)
```

```python
Q1 = df['price_per_sqft'].quantile(0.25)
Q3 = df['price_per_sqft'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
lower_limit_iqr = Q1 - 1.5 * IQR
upper_limit_iqr = Q3 + 1.5 * IQR
print(lower_limit_iqr)
print(upper_limit_iqr)
df_cleaned_iqr = df[(df['price_per_sqft'] >= lower_limit_iqr) & (df['price_per_sqft'] <= upper_limit_iqr)]
print(df_cleaned_iqr)
```

```python
from scipy.stats import zscore
df['z_score'] = zscore(df['price_per_sqft'])
df_cleaned_zscore = df[df['z_score'].abs() <= 3]
```

```python
print(df_cleaned_zscore)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.boxplot(df['price_per_sqft'])
plt.title('Original Data')
plt.subplot(2, 2, 2)
sns.boxplot(df_cleaned_mean_std['price_per_sqft'])
plt.title('After Mean and STD Outlier Removal')
plt.subplot(2, 2, 3)
sns.boxplot(df_cleaned_percentile['price_per_sqft'])
plt.title('After Percentile Outlier Removal')
plt.subplot(2, 2, 4)
sns.boxplot(df_cleaned_iqr['price_per_sqft'])
plt.title('After IQR Outlier Removal')
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import skew, kurtosis
```

```python
plt.figure(figsize=(8, 6))
```

```python
sns.histplot(df['price_per_sqft'], kde=True)
```

```python
plt.title('Price per sqft - Histogram')
plt.show()
```

```python
print(f"Skewness: {skew(df['price_per_sqft'])}")
print(f"Kurtosis: {kurtosis(df['price_per_sqft'])}")
```

```python
df['price_per_sqft_log'] = np.log(df['price_per_sqft'] + 1)
plt.figure(figsize=(8, 6))
sns.histplot(df['price_per_sqft_log'], kde=True)
plt.title('Log Transformed Price per sqft - Histogram')
plt.show()
print(f"Skewness after log transformation: {skew(df['price_per_sqft_log'])}")
print(f"Kurtosis after log transformation: {kurtosis(df['price_per_sqft_log'])}")
```

```python

df_numeric = df.select_dtypes(include=[np.number])
corr_matrix = df_numeric.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
